## **Lab: Working with Git/GitHub - Pull Requests**

### **Objective:**
In this lab, you will learn how to:
- Fork a repository
- Create a branch
- Make changes and commit them
- Create a pull request (PR)
- Merge a PR into the main branch

## **Step 1: Fork the Repository**

1. Student A will create a repository on GitHub and share the link with the other student (Student B). Use the starter code provided.
2. Student B will go to the shared repository link.
3. Student B will Fork the repository by clicking the "Fork" button in the upper right corner to create a copy of the repository under their GitHub account.

## **Step 2: Clone Your Forked Repository**

1. Student B should clone the forked repository to their local machine:

    ```bash
    git clone <<repo url here>>
    ```

2. Navigate to the repository directory:

    ```bash
    cd repository
    ```

## **Step 3: Create a Branch**

1. Student B will create a new branch to make changes:

    ```bash
    git checkout -b <<branch name here>>
    ```

## **Step 4: Make Changes and Commit Them**

1. Student B opens the `main.py` and will complete the functions.
2. Save the changes.
3. Add the changes to the staging area:

    ```bash
    git add main.py
    ```

4. Commit the changes with a meaningful commit message:

    ```bash
    git commit -m "add your commit message here"
    ```

## **Step 5: Push Your Branch to GitHub**

1. Student B pushes their branch to their GitHub repository:

    ```bash
    git push origin <<branch name here>> [//]: # "Note: adding ... origin <<branch name>> only needs to happen once when pushing a new branch"
    ```

## **Step 6: Create a Pull Request (PR)**

1. Student B goes to their forked repository on GitHub.
2. Clicks on the "Compare & pull request" button to create a PR back to the original repository (owned by Student A).
3. Provides a meaningful title and description for the PR.
4. Click "Create pull request."


## **Step 7: Accept Pull Request**

1. Student A will review and accept the pull request, merging the PR on GitHub.
2. Student A will the merge the changes to their local repository.

## **Step 8: Repeat the Process**

1. Students A and B will reverse roles and work through the previous steps.

## **Step 9: Review and Discuss**

- Both students discuss what they learned about working with Git and GitHub.
